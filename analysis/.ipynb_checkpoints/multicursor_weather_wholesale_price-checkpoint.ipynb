```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import matplotlib.patheffects as path_effects

pd.set_option('max_columns', None)
```

```python
def rem_ws_col(df):
    df.columns = df.columns.str.strip()
    return df

def rem_ws_val(df):
    df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
    return df



folder = 'finals_2021_1'
path = '/home/danguyen/data/powertac/finals_2021/{}/analysis'.format(folder)
cwd = Path(path)
broker_mkt_prices = pd.read_csv(cwd/"{}.broker-market-prices.csv".format(folder), delimiter=",", decimal=".")
broker_mkt_prices = rem_ws_col(broker_mkt_prices)
weather = pd.read_csv(cwd/"{}.production-consumption-weather.csv".format(folder), delimiter=",", decimal=".")
weather = rem_ws_col(weather)


folder = 'finals_2021_2'
path = '/home/danguyen/data/powertac/finals_2021/{}/analysis'.format(folder)
cwd = Path(path)
broker_mkt_prices2 = pd.read_csv(cwd/"{}.broker-market-prices.csv".format(folder), delimiter=",", decimal=".")
broker_mkt_prices2 = rem_ws_col(broker_mkt_prices2)
weather2 = pd.read_csv(cwd/"{}.production-consumption-weather.csv".format(folder), delimiter=",", decimal=".")
weather2 = rem_ws_col(weather2)

folder = 'finals_2021_3'
path = '/home/danguyen/data/powertac/finals_2021/{}/analysis'.format(folder)
cwd = Path(path)
broker_mkt_prices3 = pd.read_csv(cwd/"{}.broker-market-prices.csv".format(folder), delimiter=",", decimal=".")
broker_mkt_prices3 = rem_ws_col(broker_mkt_prices3)
weather3 = pd.read_csv(cwd/"{}.production-consumption-weather.csv".format(folder), delimiter=",", decimal=".")
weather3 = rem_ws_col(weather3)

broker_mkt_prices = broker_mkt_prices[['ts', 'default broker price', 'CrocodileAgent2020 price', 'COLDPOWER2021 price', 'VidyutVanika price', 'IS3 price', 'TUC_TAC price', 'Mertacor2021 price']]
broker_mkt_prices2 = broker_mkt_prices2[['ts', 'default broker price', 'CrocodileAgent2020 price', 'COLDPOWER2021 price', 'VidyutVanika price', 'IS3 price', 'TUC_TAC price', 'Mertacor2021 price']]
broker_mkt_prices3 = broker_mkt_prices3[['ts', 'default broker price', 'CrocodileAgent2020 price', 'COLDPOWER2021 price', 'VidyutVanika price', 'IS3 price', 'TUC_TAC price', 'Mertacor2021 price']]

melted_prices = pd.melt(broker_mkt_prices, id_vars=['ts'], value_vars=[col for col in broker_mkt_prices.columns if 'price' in col], 
                    var_name='broker', value_name='price')
melted_prices2 = pd.melt(broker_mkt_prices2, id_vars=['ts'], value_vars=[col for col in broker_mkt_prices2.columns if 'price' in col], 
                    var_name='broker', value_name='price')
melted_prices3 = pd.melt(broker_mkt_prices3, id_vars=['ts'], value_vars=[col for col in broker_mkt_prices3.columns if 'price' in col], 
                    var_name='broker', value_name='price')

melted_prices['broker'] = melted_prices['broker'].str.replace(' price', '')
melted_prices2['broker'] = melted_prices2['broker'].str.replace(' price', '')
melted_prices3['broker'] = melted_prices3['broker'].str.replace(' price', '')

final_prices = melted_prices.append(melted_prices2).append(melted_prices3)

final_weather = weather.append(weather2).append(weather3)
final_weather = final_weather[['slot', 'production', 'consumption', 'temp', 'wind', 'cloud']]
final_weather.columns =['ts', 'production', 'consumption', 'temp', 'wind', 'cloud']
```

```python
final_prices_agg = final_prices.groupby(['ts', 'broker'], as_index=False).agg({'price':['mean','std']})
final_prices_agg.columns = final_prices_agg.columns.get_level_values(0)
final_prices_agg.columns = ['ts', 'broker', 'mean', 'std']
final_prices_agg = final_prices_agg.fillna(0)
final_prices_agg['upper_bound'] = final_prices_agg['mean'] + 2 * final_prices_agg['std']
final_prices_agg['lower_bound'] = final_prices_agg['mean'] - 2 * final_prices_agg['std']
final_prices_agg
```

```python
final_prices_agg[final_prices_agg['broker'] == 'COLDPOWER2021']
```

```python
final_weather_agg = final_weather.groupby(['ts'], as_index=False).agg({'production':['mean','std'], 'consumption':['mean','std'], 'wind':['mean','std'], 'cloud':['mean','std']})
final_weather_agg.columns = final_weather_agg.columns.get_level_values(0)
final_weather_agg.columns = ['ts', 'prod_mean', 'prod_std', 'cons_mean', 'cons_std', 'wind_mean', 'wind_std', 'cloud_mean', 'cloud_std']
final_weather_agg = final_weather_agg.fillna(0)
final_weather_agg['prod_ub'] = final_weather_agg['prod_mean'] + 2 * final_weather_agg['prod_std']
final_weather_agg['prod_lb'] = final_weather_agg['prod_mean'] - 2 * final_weather_agg['prod_std']

final_weather_agg['cons_ub'] = final_weather_agg['cons_mean'] + 2 * final_weather_agg['cons_std']
final_weather_agg['cons_lb'] = final_weather_agg['cons_mean'] - 2 * final_weather_agg['cons_std']

final_weather_agg['wind_ub'] = final_weather_agg['wind_mean'] + 2 * final_weather_agg['wind_std']
final_weather_agg['wind_lb'] = final_weather_agg['wind_mean'] - 2 * final_weather_agg['wind_std']

final_weather_agg['clound_ub'] = final_weather_agg['cloud_mean'] + 2 * final_weather_agg['cloud_std']
final_weather_agg['cloud_lb'] = final_weather_agg['cloud_mean'] - 2 * final_weather_agg['cloud_std']

final_weather_agg
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

fig = make_subplots(rows=4, cols=1,  shared_xaxes=True)

fig.add_trace(go.Scatter(x=final_prices_agg[final_prices_agg['broker'] == 'COLDPOWER2021']['ts'], y=final_prices_agg[final_prices_agg['broker'] == 'COLDPOWER2021']['mean'], name="Mean Price",mode='lines',
                         line_color='deepskyblue'), row=1, col=1)

fig.add_trace(go.Scatter(x=final_prices_agg[final_prices_agg['broker'] == 'COLDPOWER2021']['ts'], y=final_prices_agg[final_prices_agg['broker'] == 'COLDPOWER2021']['upper_bound'], name="Upper Bound",
        mode='lines',
        marker=dict(color="#444"),
        line=dict(width=0),
        showlegend=False
    ), row=1, col=1)
fig.add_trace(go.Scatter(x=final_prices_agg[final_prices_agg['broker'] == 'COLDPOWER2021']['ts'], y=final_prices_agg[final_prices_agg['broker'] == 'COLDPOWER2021']['lower_bound'], name="Lower Bound",
        marker=dict(color="#444"),
        line=dict(width=0),
        mode='lines',
        fillcolor='rgba(68, 68, 68, 0.3)',
        fill='tonexty',
        showlegend=False), row=1, col=1)

fig.update_layout(yaxis1 = dict(range=[-80, 10]))
fig.update_layout(yaxis3 = dict(range=[-70, -30]))


fig.add_trace(go.Scatter(x=final_weather_agg['ts'], y=final_weather_agg['prod_mean'], name="Production Mean",
                         line_color='green'), row=2, col=1)
fig.add_trace(go.Scatter(x=final_weather_agg['ts'], y=final_weather_agg['prod_ub'], name="Upper Bound",
        mode='lines',
        marker=dict(color="#444"),
        line=dict(width=0),
        showlegend=False
    ), row=2, col=1)
fig.add_trace(go.Scatter(x=final_weather_agg['ts'], y=final_weather_agg['prod_lb'], name="Lower Bound",
        marker=dict(color="#444"),
        line=dict(width=0),
        mode='lines',
        fillcolor='rgba(68, 68, 68, 0.3)',
        fill='tonexty',
        showlegend=False), row=2, col=1)

fig.add_trace(go.Scatter(x=final_weather_agg['ts'], y=final_weather_agg['cons_mean'], name="Consumption Mean",
                         line_color='red'), row=3, col=1)
fig.add_trace(go.Scatter(x=final_weather_agg['ts'], y=final_weather_agg['cons_ub'], name="Upper Bound",
        mode='lines',
        marker=dict(color="#444"),
        line=dict(width=0),
        showlegend=False
    ), row=3, col=1)
fig.add_trace(go.Scatter(x=final_weather_agg['ts'], y=final_weather_agg['cons_lb'], name="Lower Bound",
        marker=dict(color="#444"),
        line=dict(width=0),
        mode='lines',
        fillcolor='rgba(68, 68, 68, 0.3)',
        fill='tonexty',
        showlegend=False), row=3, col=1)

fig.add_trace(go.Scatter(x=final_weather_agg['ts'], y=final_weather_agg['wind_mean'], name="Wind Speed Mean",
                         line_color='red'), row=4, col=1)
fig.add_trace(go.Scatter(x=final_weather_agg['ts'], y=final_weather_agg['wind_ub'], name="Upper Bound",
        mode='lines',
        marker=dict(color="#444"),
        line=dict(width=0),
        showlegend=False
    ), row=4, col=1)
fig.add_trace(go.Scatter(x=final_weather_agg['ts'], y=final_weather_agg['wind_lb'], name="Lower Bound",
        marker=dict(color="#444"),
        line=dict(width=0),
        mode='lines',
        fillcolor='rgba(68, 68, 68, 0.3)',
        fill='tonexty',
        showlegend=False), row=4, col=1)




fig.update_layout(title='Price per kWh on wholesale market of COLDPOWER2021',
    yaxis_title='Price per kWh',
                  
                  xaxis4_title = 'Timeslot',
                  xaxis4_rangeslider_visible=True,
                  height=600)

fig.update_layout(#height=600, width=1400,
    hovermode = "x unified",
    legend_traceorder="normal")
fig.update_traces(xaxis='x4')

fig.show()
```

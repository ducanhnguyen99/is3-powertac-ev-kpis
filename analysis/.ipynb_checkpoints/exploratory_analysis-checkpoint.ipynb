```python
# remove leading and trailing whitespaces in column names and values

def rem_ws_col(df):
    df.columns = df.columns.str.strip()
    return df

def rem_ws_val(df):
    df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
    return df

tf_transactions = rem_ws_col(tf_transactions)
balancing_actions = rem_ws_col(balancing_actions)

tf_transactions[['broker-name', 'tariff-type', 'transaction-type']].astype(str)
tf_transactions = rem_ws_val(tf_transactions)
```

```python
def pivot_broker_accounting(broker_accounting):
    # pivot balancing actions longer to have one broker column with the respective loads and payments
    iterations = range(int((len(broker_accounting.columns)-3)/18))

    pivoted_broker_accounting = pd.DataFrame()

    # for the i the number of brokers pivot the required columns
    for i in iterations:
        common_columns = broker_accounting.iloc[:, 0]
        broker_columns = broker_accounting.iloc[:, 3+i*18:(i+1)*18+3]
        broker_columns.columns = ['broker', 'ttx-sc', 'ttx-sd', 'ttx-uc', 'ttx-ud', 'mtx-c', 'mtx-d', 'btx-c', 'btx-d', 'dtx-c', 'dtx-d', 'ctx-c', 'ctx-d', 'bce-c', 'bce-d', 'bank-c', 'bank-d', 'cash']
        df_merge = pd.concat([common_columns,broker_columns], axis=1)
        pivoted_broker_accounting = pd.concat([pivoted_broker_accounting, df_merge])

    return pivoted_broker_accounting
                                
```

```python
new_df_credit = new_df[['broker', 'ttx-sc', 'ttx-uc', 'mtx-c', 'btx-c']].copy()
new_df_credit['credit'] = new_df_credit.iloc[:, range(1,5)].sum(axis=1)
new_df_debit = new_df[['broker', 'ttx-sd', 'ttx-ud', 'mtx-d', 'btx-d']].copy()
new_df_debit['debit'] = new_df_debit.iloc[:, range(1,5)].sum(axis=1)

new_df_credit['ttx-sc'] = new_df_credit['ttx-sc']/new_df_credit['credit']
new_df_credit['ttx-uc'] = new_df_credit['ttx-uc']/new_df_credit['credit']
new_df_credit['mtx-c'] = new_df_credit['mtx-c']/new_df_credit['credit']
new_df_credit['btx-c'] = new_df_credit['btx-c']/new_df_credit['credit']

new_df_debit['ttx-sd'] = new_df_debit['ttx-sd']/new_df_debit['debit']
new_df_debit['ttx-ud'] = new_df_debit['ttx-ud']/new_df_debit['debit']
new_df_debit['mtx-d'] = new_df_debit['mtx-d']/new_df_debit['debit']
new_df_debit['btx-d'] = new_df_debit['btx-d']/new_df_debit['debit']
```

```python
credit_melt = pd.melt(new_df_credit, id_vars=['broker'], value_vars=['ttx-sc', 'ttx-uc', 'mtx-c', 'btx-c'], var_name='cost_type', value_name='cost')
credit_melt.fillna(0)

sns.set_style("whitegrid")
fig, axes = plt.subplots(2,2,figsize=(16,16))
axes[0,0].pie(credit_melt[credit_melt['broker'] == 'IS3']['cost'], labels=credit_melt[credit_melt['broker'] == 'IS3' ]['cost_type'], autopct='%1.1f%%', startangle=90)
axes[0,1].pie(credit_melt[credit_melt['broker'] == 'TUC_TAC']['cost'], labels=credit_melt[credit_melt['broker'] == 'TUC_TAC']['cost_type'], autopct='%1.1f%%', startangle=90)
axes[1,0].pie(credit_melt[credit_melt['broker'] == 'Mertacor2021']['cost'], labels=credit_melt[credit_melt['broker'] == 'Mertacor2021']['cost_type'], autopct='%1.1f%%', startangle=90)
axes[1,1].pie(credit_melt[credit_melt['broker'] == 'COLDPOWER2021']['cost'], labels=credit_melt[credit_melt['broker'] == 'CrocodileAgent2020']['cost_type'], autopct='%1.1f%%', startangle=90)
axes[0,0].set(xlabel="IS3")
axes[0,1].set(xlabel="TUC_TAC")
axes[1,0].set(xlabel="Mertacor2021")
axes[1,1].set(xlabel="COLDPOWER2021")

axes[0,0].legend()
axes[0,1].legend()
axes[1,0].legend()
axes[1,1].legend()
fig.suptitle('Share of Brokers Revenue in Retail/Balancing/Wholesale Market')

plt.show()
```

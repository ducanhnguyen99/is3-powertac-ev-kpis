## Comparing groups of games

```python
def pivot_ba(ba):
    iterations = range(int((len(ba.columns)-6)/7))

    pivoted_balancing_actions = pd.DataFrame()

    # for the i the number of brokers pivot the required columns
    for i in iterations:
        common_columns = ba.iloc[:, [0,3]]
        broker_columns = ba.iloc[:, 6+i*7:(i+1)*7+6]
        broker_columns.columns = ['broker', 'netLoad', 'regOffered', 'regUsed', 'baseCost', 'p1', 'p2']
        df_merge = pd.concat([common_columns,broker_columns], axis=1)
        pivoted_balancing_actions = pd.concat([pivoted_balancing_actions, df_merge])
    pivoted_balancing_actions['regUsed'] = abs(pivoted_balancing_actions['regUsed'])
    pivoted_balancing_actions['totalImbalance'] = abs(pivoted_balancing_actions['totalImbalance'])
    
    return pivoted_balancing_actions
```

```python
path = '/data/passive/powertac/games/finals_2020_11/extractions/finals_2020_11_1_broker-balancing-actions.csv'
ba1 = pd.read_csv(path, delimiter=";", skipinitialspace=True)
path = '/data/passive/powertac/games/finals_2020_11/extractions/finals_2020_11_2_broker-balancing-actions.csv'
ba2 = pd.read_csv(path, delimiter=";", skipinitialspace=True)
path = '/data/passive/powertac/games/finals_2020_11/extractions/finals_2020_11_3_broker-balancing-actions.csv'
ba3 = pd.read_csv(path, delimiter=";", skipinitialspace=True)

pivot_ba1 = pivot_ba(ba1)
pivot_ba2 = pivot_ba(ba2)
pivot_ba3 = pivot_ba(ba3)

final_ba1 = pivot_ba1[['ts', 'totalImbalance', 'regUsed']].groupby(by = ['ts', 'totalImbalance']).sum().reset_index()
final_ba2 = pivot_ba2[['ts', 'totalImbalance', 'regUsed']].groupby(by = ['ts', 'totalImbalance']).sum().reset_index()
final_ba3 = pivot_ba3[['ts', 'totalImbalance', 'regUsed']].groupby(by = ['ts', 'totalImbalance']).sum().reset_index()

merged_ba = pd.concat([final_ba1, final_ba2, final_ba3], axis=0)
merged_ba['ts2'] = merged_ba['ts'] % 24
merged_ba['percReduced'] = merged_ba['regUsed'] / merged_ba['totalImbalance']*100
```

```python
path = '/data/passive/powertac/games/finals_2020_11/extractions/finals_2020_11_46_broker-balancing-actions.csv'
ba46 = pd.read_csv(path, delimiter=";", skipinitialspace=True)
path = '/data/passive/powertac/games/finals_2020_11/extractions/finals_2020_11_47_broker-balancing-actions.csv'
ba47 = pd.read_csv(path, delimiter=";", skipinitialspace=True)
path = '/data/passive/powertac/games/finals_2020_11/extractions/finals_2020_11_48_broker-balancing-actions.csv'
ba48 = pd.read_csv(path, delimiter=";", skipinitialspace=True)

pivot_ba46 = pivot_ba(ba46)
pivot_ba47 = pivot_ba(ba47)
pivot_ba48 = pivot_ba(ba48)

final_ba46 = pivot_ba46[['ts', 'totalImbalance', 'regUsed']].groupby(by = ['ts', 'totalImbalance']).sum().reset_index()
final_ba47 = pivot_ba47[['ts', 'totalImbalance', 'regUsed']].groupby(by = ['ts', 'totalImbalance']).sum().reset_index()
final_ba48 = pivot_ba48[['ts', 'totalImbalance', 'regUsed']].groupby(by = ['ts', 'totalImbalance']).sum().reset_index()

merged_ba2 = pd.concat([final_ba46, final_ba47, final_ba48], axis=0)
merged_ba2['ts2'] = merged_ba2['ts'] % 24
merged_ba2['percReduced'] = merged_ba2['regUsed'] / merged_ba2['totalImbalance']*100
```

```python
path = '/data/passive/powertac/games/finals_2020_11/extractions/finals_2020_11_302_broker-balancing-actions.csv'
ba302 = pd.read_csv(path, delimiter=";", skipinitialspace=True)
path = '/data/passive/powertac/games/finals_2020_11/extractions/finals_2020_11_304_broker-balancing-actions.csv'
ba304 = pd.read_csv(path, delimiter=";", skipinitialspace=True)
path = '/data/passive/powertac/games/finals_2020_11/extractions/finals_2020_11_307_broker-balancing-actions.csv'
ba307 = pd.read_csv(path, delimiter=";", skipinitialspace=True)

pivot_ba302 = pivot_ba(ba302)
pivot_ba304 = pivot_ba(ba304)
pivot_ba307 = pivot_ba(ba307)

final_ba302 = pivot_ba302[['ts', 'totalImbalance', 'regUsed']].groupby(by = ['ts', 'totalImbalance']).sum().reset_index()
final_ba304 = pivot_ba304[['ts', 'totalImbalance', 'regUsed']].groupby(by = ['ts', 'totalImbalance']).sum().reset_index()
final_ba307 = pivot_ba307[['ts', 'totalImbalance', 'regUsed']].groupby(by = ['ts', 'totalImbalance']).sum().reset_index()

merged_ba3 = pd.concat([final_ba302, final_ba304, final_ba307], axis=0)
merged_ba3['ts2'] = merged_ba3['ts'] % 24
merged_ba3['percReduced'] = merged_ba3['regUsed'] / merged_ba3['totalImbalance']*100

```

```python
fig, axes = plt.subplots(1, 3, sharey=True, figsize=(12, 4))

violinplot = sns.violinplot(y='percReduced', data=merged_ba, ax = axes[0])
violinplot = sns.violinplot(y='percReduced', data=merged_ba2, ax = axes[1])
violinplot = sns.violinplot(y='percReduced', data=merged_ba3, ax = axes[2])
```
